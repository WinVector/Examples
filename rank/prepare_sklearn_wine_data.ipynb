```python
# set up Python
import logging
import numpy as np
import pandas as pd
import json
from sklearn.datasets import load_wine
from sklearn.linear_model import LogisticRegression
```

```python
# load features data
wine_data = load_wine()
wine_frame = pd.DataFrame(wine_data.data)
wine_frame.columns = wine_data.feature_names

wine_frame
```

```python
# build preference we are trying to learn
wine_class = wine_data.target
taste_model = LogisticRegression(max_iter=1000)
taste_model.fit(wine_frame, wine_class==1)
beta = taste_model.coef_[0]

beta
```

```python
wine_scores = pd.DataFrame({'score': wine_frame @ beta})

wine_scores
```

```python
# save 
wine_frame.to_csv('sklearn_wine_example_features.csv', index=False)
wine_scores.to_csv('sklearn_wine_example_scores.csv', index=False)
```
